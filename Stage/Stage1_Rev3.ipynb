## EDA (Exploratory Data Analysis)

### Dataset Overview atau Deskripsi Data

```python
work_hours = df_out_time.iloc[:, 1:] - df_in_time.iloc[:, 1:]
```

```python
work_hours = (df_out_time.iloc[:, 1:] - df_in_time.iloc[:, 1:]) / pd.Timedelta(hours=1)
```

```python
total_work_hours = work_hours.sum(axis=1)
```

```python
average_work_hours = work_hours.mean(axis=1)
```

```python
# convert datetime
df_in_time.iloc[:, 1:] = df_in_time.iloc[:, 1:].apply(pd.to_datetime, errors='coerce')
df_out_time.iloc[:, 1:] = df_out_time.iloc[:, 1:].apply(pd.to_datetime, errors='coerce')

# ubah ke long format
in_long = df_in_time.melt(id_vars=['Unnamed: 0'],
                     var_name='date',
                     value_name='in_time')

out_long = df_out_time.melt(id_vars=['Unnamed: 0'],
                       var_name='date',
                       value_name='out_time')

# merge
df = in_long.merge(out_long, on=['Unnamed: 0', 'date'])

# drop null
df = df.dropna(subset=['in_time', 'out_time'])

# hitung jam kerja per hari
df['work_hours'] = (df['out_time'] - df['in_time']).dt.total_seconds() / 3600

# buang data error
df = df[df['work_hours'] >= 0]

# tentukan overwork (>= 9 jam)
df['is_overwork'] = df['work_hours'] >= 9

# hitung jumlah hari overwork per karyawan
overwork_days = df.groupby('Unnamed: 0')['is_overwork'].sum()
overwork_days = overwork_days.reset_index(drop=True)
print(overwork_days)
```

```python
df_in_out = pd.concat(
    [total_work_hours, average_work_hours, overwork_days],
    axis=1
)

df_in_out.columns = ['total_work_hours', 'average_work_hours', 'overwork_days']
df_in_out = df_in_out.reset_index()
```

```python
df_in_out['overwork'] = df_in_out['total_work_hours'].apply(
    lambda x: 'Yes' if x > 2080 else 'No'
)
```

```python
# drop kolom index
df_in_out = df_in_out.drop(columns=['index'])

# tambah EmployeeID mulai dari 1
df_in_out.insert(0, 'EmployeeID', range(1, len(df_in_out) + 1))
```

### Univariate Analysis

```python
import math

features = categoricals_df

n_cols = 2  # lebih sedikit kolom biar lega
n_rows = math.ceil(len(features) / n_cols)

fig, axes = plt.subplots(n_rows, n_cols, figsize=(15, 6 * n_rows))

for i, ax in enumerate(axes.flatten()):
    if i < len(features):
        col = features[i]
        order = final_df[col].value_counts().index

        sns.countplot(y=final_df[col], order=order, color='#1f3b73', ax=ax)
        ax.set_title(col, fontsize=11)
    else:
        ax.set_visible(False)

plt.tight_layout()
plt.show()
```

Based on the results of the univariate analysis, the following observations were identified:
- Employee retention exceeds attrition
- Most employees rarely engage in business travel
- The largest share of employees work in the Research & Development department
- Life Sciences is the most common educational background
- The workforce is predominantly male
- Sales Executive is the most prevalent job role
- The majority of employees are married
- Most employees do not work overtime

### Bivariate Analysis

```python
import math

# rebuild dari final_df
categoricals_df = final_df.select_dtypes(include='object').columns

# optional: drop target
if 'Attrition' in categoricals_df:
    categoricals_df = categoricals_df.drop('Attrition')

features = categoricals_df

n_cols = 2
n_rows = math.ceil(len(features) / n_cols)

fig, axes = plt.subplots(n_rows, n_cols, figsize=(15, 6 * n_rows))

for i, ax in enumerate(axes.flatten()):
    if i < len(features):
        col = features[i]

        sns.countplot(
            data=final_df,
            y=col,
            hue='Attrition',
            order=final_df[col].value_counts().index,
            palette=['#1f3b73', '#7fb3ff'],
            ax=ax
        )

        ax.set_title(f'{col} vs Attrition', fontsize=11)
        ax.set_xlabel('Count')

        for container in ax.containers:
            ax.bar_label(container, fontsize=8)

    else:
        ax.set_visible(False)

plt.tight_layout()
plt.show()
```

```python
import math

# rebuild categorical
categoricals_df = final_df.select_dtypes(include='object').columns

if 'Attrition' in categoricals_df:
    categoricals_df = categoricals_df.drop('Attrition')

features = categoricals_df

n_cols = 2
n_rows = math.ceil(len(features) / n_cols)

fig, axes = plt.subplots(n_rows, n_cols, figsize=(15, 6 * n_rows))

for i, ax in enumerate(axes.flatten()):
    if i < len(features):
        col = features[i]

        #  hitung percentage
        prop_df = (
            final_df
            .groupby(col)['Attrition']
            .value_counts(normalize=True)
            .rename('percentage')
            .mul(100)
            .reset_index()
        )

        sns.barplot(
            data=prop_df,
            y=col,
            x='percentage',
            hue='Attrition',
            palette=['#1f3b73', '#7fb3ff'],
            ax=ax
        )

        ax.set_title(f'{col} vs Attrition (%)', fontsize=11)
        ax.set_xlabel('Percentage (%)')

        #  label percentage
        for container in ax.containers:
            ax.bar_label(container, fmt='%.1f%%', fontsize=8)

    else:
        ax.set_visible(False)

plt.tight_layout()
plt.show()
```

For Bivariate Analysis we count with data count per column and percentage of that column, and as we can see this results in two different visualizations because they are based on different calculation approaches.

Based on the results of the bivariate analysis, the following observations were identified:
- Frequent business travel increases attrition risk, although employees who rarely travel contribute the highest number of attrition cases
- Human Resources has the highest attrition rate, while Research & Development contributes the largest number of attrition cases
- Employees with a Human Resources educational background show the highest attrition rate, while Life Sciences contributes the largest number of attrition cases
- Attrition is not significantly influenced by gender, despite higher attrition counts among male employees
- Research Director has the highest attrition rate, while Sales Executive and Research Scientist contribute the most attrition cases
- Single employees are the most at-risk group, having both the highest attrition count and rate
- Overworked employees show significantly higher attrition rates, despite non-overworked employees contributing more cases overall

While some groups contribute more to attrition in absolute numbers, the highest attrition risk is concentrated among frequent travelers, HR-related roles/backgrounds, single employees, and overworked employees

Based on the insights that we get from this analysis, we can suggest few Business Recommendation,

 - Reduce excessive business travel
    → Implement travel policies, remote collaboration tools, or rotation systems
- Prioritize retention programs in HR department
    → Conduct engagement surveys, workload balancing, and career path clarity
- Target high-risk employee segments (HR background & roles)
    → Provide tailored development programs and retention incentives
- Focus on work-life balance initiatives
    → Especially for overworked employees (workload redistribution, flexible hours)
- Develop retention strategies for single employees
    → Offer benefits aligned with lifestyle needs (career growth, flexibility, social engagement)
- Monitor critical job roles (e.g., Research Director)
    → Implement succession planning and leadership retention strategies
