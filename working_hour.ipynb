```python
import pandas as pd

in_df = pd.read_csv('in_time.csv')
out_df = pd.read_csv('out_time.csv')

# convert semua kolom tanggal jadi datetime
in_df.iloc[:, 1:] = in_df.iloc[:, 1:].apply(pd.to_datetime)
out_df.iloc[:, 1:] = out_df.iloc[:, 1:].apply(pd.to_datetime)
```

```python
work_hours = (out_df.iloc[:, 1:] - in_df.iloc[:, 1:]) / pd.Timedelta(hours=1)
```

```python
work_hours.head()
```

```python
total_work_hours = work_hours.sum(axis=1)
```

```python
average_work_hours = work_hours.mean(axis=1)
```

```python
average_work_hours.head()
```

```python
total_work_hours.head()
```

```python
# convert datetime
in_df.iloc[:, 1:] = in_df.iloc[:, 1:].apply(pd.to_datetime, errors='coerce')
out_df.iloc[:, 1:] = out_df.iloc[:, 1:].apply(pd.to_datetime, errors='coerce')

# ubah ke long format
in_long = in_df.melt(id_vars=['Unnamed: 0'],
                     var_name='date',
                     value_name='in_time')

out_long = out_df.melt(id_vars=['Unnamed: 0'],
                       var_name='date',
                       value_name='out_time')

# merge
df = in_long.merge(out_long, on=['Unnamed: 0', 'date'])

# drop null
df = df.dropna(subset=['in_time', 'out_time'])

# hitung jam kerja per hari
df['work_hours'] = (df['out_time'] - df['in_time']).dt.total_seconds() / 3600

# buang data error
df = df[df['work_hours'] >= 0]

# tentukan overwork (>= 9 jam)
df['is_overwork'] = df['work_hours'] >= 9

# hitung jumlah hari overwork per karyawan
overwork_days = df.groupby('Unnamed: 0')['is_overwork'].sum()
overwork_days = overwork_days.reset_index(drop=True)
print(overwork_days)
```

```python
df_in_out = pd.concat(
    [total_work_hours, average_work_hours, overwork_days],
    axis=1
)

df_in_out.columns = ['total_work_hours', 'average_work_hours', 'overwork_days']
df_in_out = df_in_out.reset_index()
```

```python
df_in_out
```

```python
df_in_out['overwork'] = df_in_out['total_work_hours'].apply(
    lambda x: 'Yes' if x > 2080 else 'No'
)
```

```python
df_in_out
```

```python
# drop kolom index
df_in_out = df_in_out.drop(columns=['index'])

# tambah EmployeeID mulai dari 1
df_in_out.insert(0, 'EmployeeID', range(1, len(df_in_out) + 1))
```
